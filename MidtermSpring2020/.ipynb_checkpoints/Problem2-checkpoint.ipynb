# Problem 2: Ising Model

### Part A
Finite systems with small $L$ tend to flip the overall magnetization because the entire system can tunnel between degenerate states. To maintain one single domain, we need an $L$ large enough so that the fluctuations are local and small enough to not matter. In the 2D Ising model, we use a lattice where $N$, the total number of spins, equal $L^2$, helping our simulation staying in one domain. Our code below runs the simulation at $T = 2.0$, which is below critical temperature, for various $L$, and sets $N$ to $L^2$. It then plots the average magnetization per spin as a function of Monte Carlo steps.

```python
J = 1.0 
T = 2.0  
H = 0.0  
MCSteps = 1000
L_values = [10, 20, 40, 80, 100, 120]
plt.figure(figsize=(10,6))
for L in L_values:
    N = L * L  # total number of spins
    sim = ising.Ising(J, L, N, T, H)
    sim.run(MCSteps)
    mvals = sim.get_mvals()
    plt.plot(mvals, label=f"L={L} (N={N})")
    
plt.xlabel("MC Step")
plt.ylabel("Average Magnetization per Spin")
plt.title("Magnetization vs. MC Step at T = 2.0 for various L (with N = L²)")
plt.legend()
plt.ylim(-1, 1)
plt.show()
```

We see as $L$ increases, $N$ increases too, as described by our relation above, and fluctuations and reversals decrease. For an "appropriate" MC simulation, we should take an $L$ of at least 80, preferably higher, and an $N$ which is the square of our $L$ value.

### Part B

We fix $L$ to be 100. As such, we take $N$ to be 10000. We then run the simulation at several temperatures below the critical temperature. We then fit the data to the expected power-law dependance:
$$
    m(T) = (T_C - T)^\beta
$$


```python
import sys
import os
sys.path.append(os.path.abspath("swig"))
import ising
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

# Parameters
J = 1.0
H = 0.0
MCSteps = 10000

L = 80
N = L * L

# Choose a narrow range of temperatures near T_c for the power-law behavior.
# For the 2D Ising model (J=1), T_c ~ 2.25, so we select up to that.
Tvals = np.linspace(2.1, 2.25, 10)
mAvgVals = []

# Run the simulation for each temperature
for T in Tvals:
    sim = ising.Ising(J, L, N, T, H)
    sim.run(MCSteps)
    mAvg = sim.get_mAvg()  # average magnetization per spin
    mAvgVals.append(abs(mAvg))  

mAvgVals = np.array(mAvgVals)

# Define the fit function: m(T) = A * (T_c - T)^beta.
def mag_fit(T, A, Tc, beta):
    return A * (Tc - T)**beta

# Initial guesses: A ~ 1, Tc ~ 2.27, beta ~ 0.125, from the starting ising file.
p0 = [1.0, 2.27, 0.125]

# We need to force Tc > max(Tvals).
# If our maximum temperature in Tvals is 2.25, we can impose Tc >= 2.26. This helps us avoid outlier simulations giving us very improbable values.
lower_bounds = [0, Tvals[-1] + 0.01, 0]
upper_bounds = [np.inf, np.inf, np.inf]

popt, pcov = opt.curve_fit(
    mag_fit, 
    Tvals, 
    mAvgVals, 
    p0=p0, 
    maxfev=10000, 
    bounds=(lower_bounds, upper_bounds)
)
A_fit, Tc_fit, beta_fit = popt

print("Fitted parameters:")
print("A =", A_fit)
print("T_c =", Tc_fit)
print("beta =", beta_fit)

# Plot the data and the fit.
plt.figure(figsize=(8,6))
plt.plot(Tvals, mAvgVals, 'bo', label='Simulation Data')
T_fit = np.linspace(Tvals.min(), Tc_fit, 100)
plt.plot(T_fit, mag_fit(T_fit, *popt), 'r-', 
         label=f'Fit: T_c = {Tc_fit:.3f}, beta = {beta_fit:.3f}')
plt.xlabel("Temperature")
plt.ylabel("Average Magnetization per Spin")
plt.title("Critical Behavior: m(T) Near T_c")
plt.legend()
plt.show()
```

Here we use data from the Monte Carlo simulations to estimate critical behavior of magnetization. We fit the curve to our equation, wherein we retrieve critical temperature and $\beta$.

### Part C

In this part, we get our energy per spin from our Monte Carlo simulation. We then differentiate this energy with respect to energy to find heat capacity.

```python
J = 1.0
H = 0.0
MCSteps = 10000
L = 100
N = L * L

# Temperature range
Tvals = np.linspace(1.0, 3.0, 20)
Evals = []

for T in Tvals:
    sim = ising.Ising(J, L, N, T, H)
    sim.run(MCSteps)
    E_avg = sim.get_eAvg()  # energy per spin from simulation
    Evals.append(E_avg)

Evals = np.array(Evals)

# Numerically differentiate to estimate the heat capacity: C
C = np.gradient(Evals, Tvals)

# Plot energy per spin versus temperature.
plt.figure(figsize=(8,6))
plt.plot(Tvals, Evals, 'ko-', label="Energy per spin")
plt.xlabel("Temperature")
plt.ylabel("Energy per Spin")
plt.title("Energy per Spin vs Temperature")
plt.legend()
plt.show()

# Plot heat capacity versus temperature.
plt.figure(figsize=(8,6))
plt.plot(Tvals, C, 'm.-', label="Heat Capacity ($\partial$E/$\partial$dT)")
plt.xlabel("Temperature")
plt.ylabel("Heat Capacity")
plt.title("Heat Capacity vs Temperature")
plt.legend()
plt.show()
```

We see two plots that describe what happens with varying temperatures. We see most notably that the heat capacity spikes near the critical temperature, which is what we would expect in terms of behavior of the material. The result of this and how it affects energy per spin is also visible in the first plot.
