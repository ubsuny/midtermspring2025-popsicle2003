# Cosmic Inflation

Cosmic inflation is the theory behind the expansion of the universe post its creation. This expansion is a phenomenon that is well observed, and explains many elements of cosmology like the horizon and flatness. Standard inflation is usually explained by a nearly constant vacuum behavior. For our purposes, we use the simplified Friedmann equations, with varying values of w, our equation of state parameter, to represent varying energy contents and how it dynamically affects expansion. In our model, these equations govern the universe's expansion rate through scale factor R(t).

# Friedmann Equations

Assuming a flat Universe k = 0 and setting R(t=0)=1, the Friedmann equations (in scaled units) are:

1. **First Friedmann Equation:**
   
   $$ H^2 \equiv \left(\frac{\dot{R}}{R}\right)^2 = \frac{8\pi G}{3}\,\rho $$

2. **Acceleration Equation (Second Friedmann Equation):**
   
   $$ \dot{H} + H^2 = \frac{\ddot{R}}{R} = -\frac{4\pi G}{3}(1+3w)\,\rho $$

3. **Continuity Equation:**
   
   $$ \dot{\rho} = -3H\,(1+w)\,\rho $$


In this setup, the equation-of-state parameter ùë§ (which relates pressure and density via p = ùë§ùúå) helps determine how the universe expands. Different values of ùë§ correspond to different kinds of matter-energy components:

-  ùë§ = 0: Non-relativistic matter

-  ùë§ = 1: A stiffer form of matter

- ùë§ = -1/3: Leads to linear expansion R(t) = 1 + t

Actual cosmic inflation is purportedly approximately -1.

For our first section of the analytical approach, this is how we derive the equations we use in our plots:

# Analytic Derivation for the Scale Factor R(t)

Starting from the continuity equation:

$$
\frac{\dot{\rho}}{\rho} = -3(1+w)\frac{\dot{R}}{R},
$$

Integrating assuming the normalization $\rho(R=1)=1\$ to obtain:

$$
\rho(R) = R^{-3(1+w)}.
$$

Substituting this back into the first Friedmann equation. Assuming the normalization

$$
\frac{8\pi G}{3} = 1,
$$

we have:

$$
\left(\frac{\dot{R}}{R}\right)^2 = R^{-3(1+w)}.
$$

Multiplying through by $R^2$ yields:

$$
\dot{R}^2 = R^{2-3(1+w)}.
$$

Taking the positive square root (for an expanding Universe):

$$
\dot{R} = R^{1-\frac{3}{2}(1+w)}.
$$

### General Case ($w \neq -\frac{1}{3}$)

Separate variables:

$$
\frac{dR}{dt} = R^{1-\frac{3}{2}(1+w)} \quad \Longrightarrow \quad \int R^{\frac{3}{2}(1+w)-1}\,dR = \int dt.
$$

Integrating, we obtain:

$$
\frac{R^{\frac{3}{2}(1+w)}}{\frac{3}{2}(1+w)} = t + \text{const.}
$$

After choosing the constant so that \(R(0)=1\) (and rescaling time if necessary), the solution can be written as:

$$
R(t) = \left[1 + \frac{3}{2}(1+w)t\right]^{\frac{2}{3(1+w)}}.
$$

### Special Case ($w = -\frac{1}{3}$)

When $w = -\frac{1}{3}$, the exponent in the ODE becomes:

$$
1 - \frac{3}{2}\left(1-\frac{1}{3}\right) = 1 - 1 = 0.
$$

Thus, the ODE reduces to:

$$
\dot{R} = 1,
$$

with the simple solution:

$$
R(t) = 1 + t.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def analytic_solution(t, w):
    """
    Returns the analytic solution R(t) for the Friedmann equation,
    given a constant equation-of-state parameter w.
    For w != -1/3:
      R(t) = [1 + 1.5*(1+w)*t]^(2/[3*(1+w)])
    For w = -1/3:
      R(t) = 1 + t
    """
    if np.isclose(w, -1/3):
        return 1 + t
    else:
        return (1 + 1.5*(1+w)*t)**(2/(3*(1+w)))

# Expanded time array from 0 to 50
t = np.linspace(0, 50, 500)
w_vals = [0, 1, -1/3]

# Create subplots for each case
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, w in enumerate(w_vals):
    R_a = analytic_solution(t, w)
    axes[i].plot(t, R_a, label='Analytic', color='r')
    axes[i].set_title(f"w = {w}")
    axes[i].set_xlabel("Time")
    if i == 0:
        axes[i].set_ylabel("Scale Factor R(t)")
    axes[i].legend()

plt.suptitle("Analytic Solution Plots (Time 0 to 50)")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Numerical Approach
In the following section, we solve the same Friedmann equations numerically using the RK4 approach. We use the ODE provided in the code below, and solve it using our RK4 approach

```python
def friedmann(t, R, w):
    """
    Defines the Friedmann ODE: dR/dt = R^(1 - 1.5*(1+w)).
    """
    exponent = 1 - 1.5 * (1 + w)
    return R ** (exponent)

def rk4_step(f, t, R, dt, w):
    """
    Performs one step of the RK4 integration.
    
    Parameters:
      f  : function, the ODE function f(t, R, w)
      t  : float, current time
      R  : float, current value of the scale factor R
      dt : float, time step
      w  : float, equation-of-state parameter
      
    Returns:
      The estimated value R(t+dt) using RK4.
    """
    k1 = f(t, R, w)
    k2 = f(t + dt/2, R + dt * k1 / 2, w)
    k3 = f(t + dt/2, R + dt * k2 / 2, w)
    k4 = f(t + dt, R + dt * k3, w)
    return R + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)

def rk4_integration(f, t_span, R0, w, num_steps):
    """
    Integrates the ODE using the RK4 method over a given time span.
    
    Parameters:
      f         : function, the ODE function f(t, R, w)
      t_span    : tuple, (t0, tf) giving the initial and final times
      R0        : float, initial value R(t0)
      w         : float, equation-of-state parameter
      num_steps : int, total number of time points (including the initial value)
    
    Returns:
      t_arr : numpy array, times at which the solution is computed
      R_arr : numpy array, the computed solution R(t)
    """
    t0, tf = t_span
    t_arr = np.linspace(t0, tf, num_steps)
    dt = t_arr[1] - t_arr[0]
    R_arr = np.empty(num_steps)
    R_arr[0] = R0
    for i in range(1, num_steps):
        R_arr[i] = rk4_step(f, t_arr[i-1], R_arr[i-1], dt, w)
    return t_arr, R_arr

# Integration parameters
t_span = (0, 50)
R0 = 1
num_steps = 500
w_vals = [0, 1, -1/3]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
for i, w in enumerate(w_vals):
    t_rk4, R_rk4 = rk4_integration(friedmann, t_span, R0, w, num_steps)
    axes[i].plot(t_rk4, R_rk4, 'b.-', markersize=3, label='Numerical RK4')
    axes[i].set_title(f"w = {w}")
    axes[i].set_xlabel("Time")
    if i == 0:
        axes[i].set_ylabel("Scale Factor R(t)")
    axes[i].legend()
    
plt.suptitle("Numerical Solution Using RK4 (Time 0 to 50)")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Integration parameters
t_span = (0, 50)
R0 = 1
num_steps = 500
w_vals = [0, 1, -1/3]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
for i, w in enumerate(w_vals):
    # Numerical solution using RK4
    t_rk4, R_rk4 = rk4_integration(friedmann, t_span, R0, w, num_steps)
    # Analytic solution computed on the same grid
    R_analytic = analytic_solution(t_rk4, w)
    
    axes[i].plot(t_rk4, R_rk4, 'b.-', markersize=3, label='Numerical RK4')
    axes[i].plot(t_rk4, R_analytic, 'r--', linewidth=2, label='Analytic')
    axes[i].set_title(f"w = {w}")
    axes[i].set_xlabel("Time")
    if i == 0:
        axes[i].set_ylabel("Scale Factor R(t)")
    axes[i].legend()

plt.suptitle("Comparison: Analytic vs RK4 Numerical (Time 0 to 50)")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

We see that our numerical approach, which is done through using an approximation analysis method, is quite close and fits what we expect from our analytical approach of solving the cosmic expansion behaviors, but is less refined. This is because it is after all an approximation, restricted by the steps of the RK4 method (the RK45 would yield a better fit). In terms of behaviors for different w, for w = 0, we see a decelerating expansion, representing domination by matter. For w = 1, we see an even stiffer expansion, supporting our hypothesis that it is a stiffer starting matter content. The last graphs show linear expansion, at the borderline between acceleration and deceleration.
